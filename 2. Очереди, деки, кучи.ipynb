```python
class queue:
    def __init__(self):
        self.array = []
        self.size = 0
        self.first = 0
        self.last = 0
        
    def push(self, x):
        self.array.append(x)
        self.size += 1
        self.last += 1
        
    def pop(self):
        if self.size == 0:
            return 'error'
        self.size -= 1
        self.first += 1
        return self.array[self.first-1]
        
    def front(self):
        if self.size == 0:
            return 'error'
        return self.array[self.first]
    
    def clear(self):
        self.array = []
        self.size = 0
        self.first = 0
        self.last = 0


f = open('input.txt', 'r')

inp = []
for line in f:
    inp.append(line.split())
f.close()

queue = queue()
for line in inp:
    match line[0]:
        case 'push':
            queue.push(int(line[1]))
            print("ok")
        case 'pop':
            x = queue.pop()
            print(x)
        case 'front':
            x = queue.front()
            print(x)
        case 'size':
            print(queue.size)
        case 'clear':
            queue.clear()
            print("ok")
        case 'exit':
            print("bye")
            break
```

```python
class deque:
    def __init__(self):
        self.array_right = []
        self.array_left = []
        self.size = 0
        self.first = -1
        self.last = 0
        
    def push_front(self, x):        
        if self.first >= -1:
            self.array_right.append(x)
        else:
            self.array_left[-(self.first+1) - 1] = x
        self.size += 1
        self.first += 1
        
    def push_back(self, x):
        if self.last >= 1:
            self.array_right[self.last - 1] = x
        else:
            self.array_left.append(x)
        self.size += 1
        self.last -= 1
        
    def pop_front(self):
        if self.size == 0:
            return 'error'
        
        if self.first >= 0:
            x = self.array_right[self.first]
            self.array_right.pop()
        else:
            x = self.array_left[-self.first - 1]
        self.size -= 1
        self.first -= 1
        return x
    
    def pop_back(self):
        if self.size == 0:
            return 'error'
        
        if self.last >= 0:
            x = self.array_right[self.last]
        else:
            x = self.array_left[-self.last - 1]
            self.array_left.pop()
        self.size -= 1
        self.last += 1
        return x
        
    def front(self):
        if self.size == 0:
            return 'error'
        
        if self.first >= 0:
            x = self.array_right[self.first]
        else:
            x = self.array_left[-self.first - 1]
        return x
    
    def back(self):
        if self.size == 0:
            return 'error'
        
        if self.last >= 0:
            x = self.array_right[self.last]
        else:
            x = self.array_left[-self.last - 1]
        return x
    
    def clear(self):
        self.array_right = []
        self.array_left = []
        self.size = 0
        self.first = -1
        self.last = 0


f = open('input.txt', 'r')

inp = []
for line in f:
    inp.append(line.split())
f.close()

deque = deque()
for line in inp:
#     print('line:', line)
#     print('first:', deque.first)
#     print('last:', deque.last)
#     print()
    
    match line[0]:
        case 'push_front':
            deque.push_front(int(line[1]))
            print("ok")
        case 'push_back':
            deque.push_back(int(line[1]))
            print("ok")
        case 'pop_front':
            print(deque.pop_front())
        case 'pop_back':
            print(deque.pop_back())
        case 'front':
            print(deque.front())
        case 'back':
            print(deque.back())
        case 'size':
            print(deque.size)
        case 'clear':
            deque.clear()
            print("ok")
        case 'exit':
            print("bye")
            break
```
